```
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1,4,9,16])
plt.title("FIRST PLOT")
plt.xlabel("X LABEL")
plt.ylabel("Y LABEL")
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1,5)
y = x**3

plt.subplot(1,2,1)
plt.plot([1,2,3,4],[1,4,9,16],"go")
plt.title("1st subplot")

plt.subplot(1,2,2)
plt.plot(x,y,"r^")
plt.title("2nd subplot")

plt.suptitle("My sub-plots")
plt.show()
```

```
import matplotlib.pyplot as plt
division = ["Div-A", "Div-B", "Div-C", "Div-D", "Div-E"]
division_average_marks = [70,82,73,65,68]
plt.bar(division,division_average_marks, color ='green')
plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```
import matplotlib.pyplot as plt
division = ["Div-A", "Div-B", "Div-C", "Div-D", "Div-E"]
division_average_marks = [70,82,73,65,68]
variance = [5,8,7,6,4]
plt.barh(division,division_average_marks, xerr = variance,color ='green')
plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.show()
```

```
import matplotlib.pyplot as plt
firms = ["Firm A", "Firm B", "Firm C","Firm D","Firm E"]
market_share = [20,25,15,10,20]
Explode = [0,0.1,0,0,0]
plt.pie(market_share, explode=Explode, labels=firms,shadow=True,startangle=60)
plt.axis('equal')
plt.legend(title = "List of Firms", loc ="upper left" )
plt.show()
```
